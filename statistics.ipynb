**GrÃ¡fica puntos dada una tabla de frequencia.**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
edades = [5, 5, 6, 6, 6, 7, 7, 7, 7, 8, 8, 9]
```

```python
Age = pd.Series(edades)
```

```python
Frequency = Age.value_counts()
print(Frequency)
```

```python
df = Frequency.to_frame(name = 'Frecuency')
```

```python
df
```

```python
df = df.reset_index().rename(columns={'index' : 'Edad'})
```

```python
df
```

```python
plt.scatter(df['Edad'], df['Frecuency'], alpha=0.7)
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de Edades')
plt.xticks(df['Edad'].unique())
plt.yticks(df['Frecuency'].unique())
```
